```python
# ...existing code...
from cnstats.stats import stats
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# ...existing code...
# result = stats(zbcode='A030201', datestr='', dbcode='hgnd')
df01 = stats(zbcode='A030201', datestr='2000-2025', dbcode='hgnd', as_df=True)
df02 = stats(zbcode='A030202', datestr='2000-2025', dbcode='hgnd', as_df=True)
# print(result)
# print(df.head())
# print(df)

# 横向按时间对齐合并为多列（若需纵向堆叠改为 axis=0）
# dfmix = pd.concat([df01, df02], axis=0, join='outer').sort_index()

# 若只想把现有代码复用为 df，使用下面一行；否则在后续代码中把 df 换成 dfmix
# df = dfmix
df = df01

import matplotlib.pyplot as plt
# 将字型切换为系统内置的中文字型
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 优先使用 '查询日期' 作为时间列，然后回退到通用时间列名
time_cols = ['查询日期','time','date','Date','日期','时间','year','Year']
time_col_found = None
for c in time_cols:
    if c in df.columns:
        df[c] = pd.to_datetime(df[c], errors='coerce')
        df = df.set_index(c)
        time_col_found = c
        break

# 如果索引存在且可排序，按时间排序
if df.index.name is not None and len(df.index) > 0:
    try:
        df = df.sort_index()
    except Exception:
        pass

# 选择数值列并绘图
num_cols = df.select_dtypes(include=['number']).columns.tolist()
if not num_cols:
    candidates = [col for col in df.columns if '值' in col or 'value' in col.lower() or '数' in col or '数值' in col]
    num_cols = candidates

import matplotlib.dates as mdates

plt.figure(figsize=(10,6))
if num_cols:
    for col in num_cols:
        plt.plot(df.index, df[col], label=col)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.YearLocator(base=1))      # 每年一个主刻度
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))    # 只显示年份
    ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(1,7)))  # 可选：每6个月一个次刻度
    plt.legend()
    plt.xlabel('时间')
    plt.ylabel('人口出生率')
    title_zb = None
    if '指标代码' in df.columns:
        title_zb = df['指标代码'].iloc[0] if len(df['指标代码'])>0 else None
    plt.title(f'指标 {title_zb} 时间序列' if title_zb else '时间序列')
    plt.gcf().autofmt_xdate(rotation=45)  # 旋转以防重叠
    plt.tight_layout()
    plt.show()
else:
    print('未找到数值列,DataFrame 列: ', df.columns.tolist())
```

```python
from cnstats.stats import stats
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

df01 = stats(zbcode='A030201', datestr='2000-2025', dbcode='hgnd', as_df=True)
df02 = stats(zbcode='A030202', datestr='2000-2025', dbcode='hgnd', as_df=True)
df = df02

num1 = df01.select_dtypes(include=['number']).columns.tolist()
num2 = df02.select_dtypes(include=['number']).columns.tolist()

# print(df.head())

if not num1 and not num2:
    print('两个表均未找到数值列, 列: ', df01.columns.tolist(), df02.columns.tolist())
else:
    plt.figure(figsize=(10,6))
    ax = plt.gca()
    for col in num1:
        ax.plot(df01.index, df01[col], label=f'df01 - {col}')
    for col in num2:
        ax.plot(df02.index, df02[col], label=f'df02 - {col}', linestyle='--')
    ax.xaxis.set_major_locator(mdates.YearLocator(base=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(1,7)))
    plt.legend()
    plt.xlabel('时间')
    plt.ylabel('数值')
    plt.title('人口出生率 与 人口死亡率 趋势')
    plt.gcf().autofmt_xdate(rotation=45)
    plt.tight_layout()
    plt.show()
```
