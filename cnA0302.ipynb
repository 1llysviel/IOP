```python
from cnstats.stats import stats
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df01 = stats(zbcode='A030201', datestr='2000-2025', dbcode='hgnd', as_df=True)
df02 = stats(zbcode='A030202', datestr='2000-2025', dbcode='hgnd', as_df=True)
df = df02

# 确保为 datetime（你已有），然后按查询日期排序并把其设为索引
if '查询日期' in df01.columns:
    df01['查询日期'] = pd.to_datetime(df01['查询日期'])
    df01 = df01.sort_values('查询日期').set_index('查询日期')
if '查询日期' in df02.columns:
    df02['查询日期'] = pd.to_datetime(df02['查询日期'])
    df02 = df02.sort_values('查询日期').set_index('查询日期')

num1 = df01.select_dtypes(include=['number']).columns.tolist()
num2 = df02.select_dtypes(include=['number']).columns.tolist()

# print(df.head())

if not num1 and not num2:
    print('两个表均未找到数值列, 列: ', df01.columns.tolist(), df02.columns.tolist())
else:
    plt.figure(figsize=(10,6))
    ax = plt.gca()
    for col in num1:
        ax.plot(df01.index, df01[col], label=f'df01 - {col}')
    for col in num2:
        ax.plot(df02.index, df02[col], label=f'df02 - {col}', linestyle='--')
    ax.xaxis.set_major_locator(mdates.YearLocator(base=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(1,7)))
    plt.legend()
    plt.xlabel('时间')
    plt.ylabel('数值')
    plt.title('人口出生率 与 人口死亡率 趋势')
    plt.gcf().autofmt_xdate(rotation=45)
    plt.tight_layout()
    plt.show()
```
