```python
from utils import fetch_stats, prepare_df, numeric_cols, plot_population

year = '1980-2025'

df01 = prepare_df(fetch_stats(zbcode='A030201', datestr=year, dbcode='hgnd'))
df02 = prepare_df(fetch_stats(zbcode='A030202', datestr=year, dbcode='hgnd'))
df03 = prepare_df(fetch_stats(zbcode='A030203', datestr=year, dbcode='hgnd'))

if not (numeric_cols(df01) or numeric_cols(df02) or numeric_cols(df03)):
    print('未找到数值列, 列: ',
          df01.columns.tolist() if df01 is not None else None,
          df02.columns.tolist() if df02 is not None else None,
          df01.columns.tolist() if df01 is not None else None)
else:
    plot_population([df01, df02, df03],
                    labels=['出生率', '死亡率', '自然增长率'],
                    styles=[None, '--', '-.'],
                    title='人口趋势',
                    mark_negative=True,
                    negative_marker='x',
                    negative_color='tab:red',
                    fill_negative=True,
                    fill_alpha=0.15)
```

```python
df04 = prepare_df(fetch_stats(zbcode='A030101', datestr=year, dbcode='hgnd'))
df05 = prepare_df(fetch_stats(zbcode='A020106', datestr=year, dbcode='hgnd'))


if not (numeric_cols(df04) or numeric_cols(df05)):
    print('未找到数值列, 列: ',
        df04.columns.tolist() if df04 is not None else None,
        df05.columns.tolist() if df05 is not None else None)

else:
    plot_population([df04, df05],
                    labels=['年末总人口', '人均GDP'],
                    styles=[None, '--', '-.'],
                    title='经济指标')
```
